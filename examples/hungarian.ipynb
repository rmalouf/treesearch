```python
import treesearch
```

```python
szeged_path = "/Users/rmalouf/Desktop/UD_Hungarian-Szeged/*.conllu"
query = """
MATCH {
    V [ upos="VERB" ];
    PV [ upos="PRON" ];
    V -> PV;
    PV << V;
}
"""
```

```python
for tree, match in treesearch.search(szeged_path, query):
    print(tree[match["V"]].lemma, tree[match["PV"]].lemma, tree[match["PV"]].upos)
```
